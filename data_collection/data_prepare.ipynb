```python
from Bio import SeqIO
import pandas as pd
```

```python

##combine metadata to seq
records = SeqIO.parse("clear_BetaCoV_spike.fasta", "fasta")
SeqIO.write(records, "clear_BetaCoV_spike.tab", "tab")
df = pd.read_csv("clear_BetaCoV_spike.tab",sep='\t',names=["taxa", "seq"])
df[['acession','version']] = df.taxa.str.split(".",expand=True) 
df.head
```

```python
meta = pd.read_csv('BetaCoV.csv')
meta.head
```

```python

## add seq to acession
data=pd.merge(df, meta, how='left', left_on='acession', right_on='Accession')
```

```python
data.head
```

```python
df_species=data.groupby(['Species']).count()
df_species
```

```python
data['taxa_new']=data.apply(lambda x:'%s|%s' % (x['acession'],x['GenBank_Title']),axis=1)
data.head
```

```python
## generate the meta
meta_spike = data.drop(['Accession', 'seq'], axis = 1)
meta_spike.to_csv("meta_spike.csv",index=False)
```

## prepare the fasta and metadata for Nextstrain

```python
spike_seq = data.filter(['taxa_new','seq'], axis=1)
```

```python
metacode = pd.read_csv('meta_spike_full_code.csv')
metacode.head()
```

```python
## read the code metacsv
metacode = pd.read_csv('meta_spike_full_code.csv')
data2=pd.merge(df, metacode, how='left', left_on='acession', right_on='acession')
data2.head()
```

```python
## get the fasta file
data2 = data2.dropna(subset=['Collection_Date'])
data2.head()
```

```python
spike_nextstrain = data2.filter(['taxa_new','seq'], axis=1)
spike_nextstrain.head
```

```python
## to fasta
spike_nextstrain.to_csv("spike_nextstrain.tab", sep="\t",index = False,header=False)
## covert csv to fasta
spike_fasta = SeqIO.parse("spike_nextstrain.tab", "tab")
SeqIO.write(spike_fasta, "spike_nextstrain.fasta", "fasta")
```
