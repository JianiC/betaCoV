```python
from opencage.geocoder import OpenCageGeocode
import pandas as pd
```

```python
metadata=pd.read_csv("TX/nextstrain/data/metadata.tsv",sep= "\t")
meta=metadata.groupby(['location']).count()
meta = meta.reset_index()
meta['state'] = 'TX'
meta = meta[["location", "state"]]
meta
```

```python
list_lat = []   # create empty lists
list_long = []
for index, row in meta.iterrows(): # iterate over rows in dataframe

    City = row['location']
    State = row['state']       
    query = str(City)+','+str(State)

    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']

    list_lat.append(lat)
    list_long.append(long)

# create new columns from lists    

meta['lat'] = list_lat   

meta['lon'] = list_long
```
