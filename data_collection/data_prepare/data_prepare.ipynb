## Data filter:
Date Guessing: for sequence without month and day information, the mid of the month and the mid of the year were used

Filter CoV19: CoV19 collected after Feb29 were removed

```python
## parse seq data
from Bio import SeqIO
```

### replace " " to "_" in fasta file


```python
BetaCoV_seq = SeqIO.parse("sequences2.fasta", "fasta")
SeqIO.write(BetaCoV_seq, "BetaCoV_seq.tab", "tab")
seq = pd.read_csv("BetaCoV_seq.tab",sep='\t',names=["taxa", "seq"])
seq.head
```

```python
##spilt taxa
#seq[['acession','gbtitle']] = seq.taxa.str.split("|",expand=True) 
spilt_col = seq["taxa"].str.split("|", expand=True)
spilt_col
```

```python
## drop the rest of column and combine with seq
spilt_col=spilt_col.iloc[:, 0:2]
spilt_col
```

```python
seq2 = seq.merge(spilt_col, left_index=True, right_index=True)
```

```python
seq2.count()
```

```python
## filter the fasta file by WGS meta table
##rename column in seq
seq2.columns.values[2] = "Accession"
seq2.columns.values[3] = "gbtitle"
seq2
```

```python
seq2.to_csv("seq2.csv",index=False)
```

```python
## WGS fasta (>20000)
WGS_meta=pd.read_csv('BetaCoV_WGS_CoV19filter.csv')
WGS_meta
```

```python
seq2= pd.read_csv("seq2.csv")
seq2
```

```python
WGS_seq=WGS_meta.merge(seq2,left_on='Accession', right_on="Accession", how='left')
```

```python
WGS_seq
```

```python
## save WGS_seq metadata and fasta
WGS_seq_fasta=WGS_seq.iloc[:, 15:17]
WGS_seq_fasta
```

```python
## to fasta

WGS_seq_fasta.to_csv("WGS_seq_fasta.tab", sep="\t",index = False,header=False)

## covert csv to fasta
WGS_seq_fasta = SeqIO.parse("WGS_seq_fasta.tab", "tab")
SeqIO.write(WGS_seq_fasta, "BetaaCoV_WGS_CoV19filter.fasta", "fasta")
```

```python
## For nextstrain deploy: drop dateNA
WGS_seq2 = WGS_seq.dropna(subset=['Collection_Date_code'])
WGS_seq2
```

```python
## get fasta
WGS_seq2_fasta=WGS_seq2.iloc[:, 15:17]

WGS_seq2_fasta.to_csv("WGS_seq2_fasta.tab", sep="\t",index = False,header=False)

## covert csv to fasta
WGS_seq2_fasta = SeqIO.parse("WGS_seq2_fasta.tab", "tab")
SeqIO.write(WGS_seq2_fasta, "BetaCoV_WGS_nextstrain.fasta", "fasta")
```

```python
## drop seq column to save metadata
WGS_seq2_meta = WGS_seq2.drop(columns=['seq'])
WGS_seq2_meta
```

```python
WGS_seq2_meta.to_csv('BetaCoV_WGS_nextstrain.csv',index = False)
```
