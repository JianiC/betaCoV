## from nextstrain meta get country and region

```python
import pandas as pd
```

```python
## seperate country from code nextstrain metadata 
df =pd.read_csv("spike_nextstrain_meta.csv")
df.head
```

```python
df[['country','state']] = df.Geo_Location.str.split(":",expand=True)
df
```

```python

# import the dictionary 
dic=pd.read_csv("country-code.csv",index_col=0, squeeze=True).to_dict()

## do translate
df['region']=df['country'].map(dic)
df
```

```python
df.to_csv("spike_nextstrain_meta2.csv", index=False)
```
