```python
def irma_seq(species, fasta_file):
    df=pd.read_excel(species)
    df_WGS=df[df['Length']>20000]
    BetaCoV_seq = SeqIO.parse("BetaCoV.fasta", "fasta")
    SeqIO.write(BetaCoV_seq, "BetaCoV_seq.tab", "tab")
    seq = pd.read_csv("BetaCoV_seq.tab",sep='\t',names=["taxa", "seq"])
    seq[['id','species']] = seq.taxa.str.split("|",expand=True) 
    seq[['Accession','version']] = seq.id.str.split(".",expand=True) 
    fasta=pd.merge(df_WGS, seq, how='left', left_on='Accession', right_on='Accession')
    fasta=fasta.filter(['taxa','seq'], axis=1)
    fasta.to_csv("temp.tab", sep="\t",index = False,header=False)
    temp = SeqIO.parse("temp.tab", "tab")
    SeqIO.write(temp,  fasta_file, "fasta")
    os.remove("temp.tab")
    print(str(len(fasta.index))+" unique records saved to " + fasta_file)
```
