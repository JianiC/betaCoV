```python
import pandas as pd
from Bio import SeqIO
```

```python
df = pd.read_csv('BetaCoV.csv')

df_species=df.groupby(['Species']).count()
df_species
```

```python
df_species.to_csv("species.csv")
```

```python
## get HCoV_OC43
OC43=pd.read_excel('HCoV-OC43.xlsx')
OC43
```

```python
OC43_WGS=OC43[OC43['Length']>20000]
OC43_WGS
```

```python
BetaCoV_seq = SeqIO.parse("sequences.fasta", "fasta")
SeqIO.write(BetaCoV_seq, "BetaCoV_seq.tab", "tab")
seq = pd.read_csv("BetaCoV_seq.tab",sep='\t',names=["taxa", "seq"])
seq
```

```python


## add seq to acession
df1=pd.merge(OC43_WGS, seq, how='left', left_on='Accession', right_on='taxa')
df1
```

```python
OC43_seq = df1.filter(['taxa','seq'], axis=1)
## to fasta
OC43_seq.to_csv("OC43.tab", sep="\t",index = False,header=False)
## covert csv to fasta
OC43_seq = SeqIO.parse("OC43.tab", "tab")
SeqIO.write(OC43_seq, "OC43.fasta", "fasta")
```
