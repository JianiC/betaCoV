```python
import pandas as pd
```

```python
ano = pd.read_csv('BetaCoV_Annotations.csv')
```

```python
CDS = ano.loc[ano['Type'] == 'CDS']
CDS_name = CDS.groupby(['Name']).count()
CDS_name
```

```python
CDS_name=CDS_name.loc[pd.to_numeric(CDS_name['Sequence Name']) >= 3]
CDS_name
```
